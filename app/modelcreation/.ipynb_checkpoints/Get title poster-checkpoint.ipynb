```python
import requests
import re
from bs4 import BeautifulSoup
from IPython.display import display

reftitleid = 'tt0055254'
reftitle = 'One Hundred and One Dalmatians'

url = 'https://www.imdb.com/title/tt0055254/'
titleString = '{} Poster'.format(reftitle)
titleString
```

```python
response = requests.get(url).content
response
```

```python
soup = BeautifulSoup(response, 'html.parser')
soup
```

```python
imgs = []

for link in soup.find_all('img', title=titleString):
    imgs.append(link.get('src'))

imgs
```
